```python
import pandas as pd
```

```python
data = pd.read_csv(r'./heart.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
data.columns.values
labels = ['age','oldpeak', 'thalachh','trtbps','thall']
sizes = [data['age'].mean(), data['oldpeak'].mean(), data['thalachh'].mean(),data['trtbps'].mean(),data['thall'].mean()]
colors = ['Red','Blue','Green','Yellow','pink']
explode = (0.1,0,0,0,0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Average data')
```

```python
import numpy as np
import seaborn as sns
```

```python
h = data.iloc[1:21, 0]
y_pos = np.arange(len(h))
v = range(1, 21)
plt.bar(y_pos, h, align='center', alpha=0.5)
plt.xticks(y_pos, v)
plt.ylabel('Value')
plt.xlabel("Age")
plt.title('Age Barplot')
plt.show()
```

```python
hist = sns.histplot(data=data, x="age")
hist.set(xlabel ="Age", ylabel = "Value", title ='Histogram for Age')
```

```python
h = data.iloc[0:16, 0]
v = data.iloc[0:16, 2]
plt.plot(h, label='age', marker='*', linestyle='dotted')
plt.plot(v, label='restecg', marker='o', linestyle='dashed')
plt.ylim(0, 60)
plt.legend()
plt.ylabel("Value Range")
plt.xlabel("Days")
plt.show()
```

```python
s=data.iloc[0:,0]
t=data.iloc[0:,3]
plt.scatter(s,t)
plt.title("Age Vs trtbps")
plt.xlabel("Age")
plt.ylabel("trtbps")
plt.show()
```

```python
data.boxplot(by='age',column=['trtbps'], grid=False)
plt.show()
```
